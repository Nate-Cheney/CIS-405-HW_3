```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

Write the code to display the names (not the codes) of the top 10 airports that have the highest number of outgoing flights. Use the datasets from the class program. 

```python
# Load airport data
airport_cols = ["ID", "Name", "City", "Country", "IATA", "ICAO", "Lat", "Lon", "Alt", "TZ", "DST", "Tz_Type", "Source"]
airports = pd.read_csv("https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat", names=airport_cols, index_col=0, na_values="\\N")
airports.head()
```

```python
# Load route data
route_cols = ["Airline", "Airline_ID", "Source_Airport", "Source_ID", "Dest_Airport", "Dest_ID", "Codeshare", "Stops", "Equip"]
routes = pd.read_csv("https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat", names=route_cols, na_values="\\N")
routes.head()
```

```python
# Clean: Remove routes where we don"t have the IATA code
routes = routes.dropna(subset=["Source_Airport", "Dest_Airport"])
```

```python
# Build multi_directional graph
G = nx.from_pandas_edgelist(routes, source="Source_Airport", target="Dest_Airport", create_using=nx.MultiDiGraph())
```

```python
# Calculate out-degrees
out_degrees = G.out_degree() 

# Sort the nodes by their out-degree and slice top-10
top_10_airports = sorted(out_degrees, key=lambda x: x[1], reverse=True)[:10]

print("Top 10 Airports by Outgoing Flights:")
for rank, (airport, flight_count) in enumerate(top_10_airports, 1):
    print(f"{rank}. {airport}: {flight_count} flights")
```
